### Alpha And Zorder in Plot Function

```python
# alpha  which is used adjust the transparency of the plot
# dfault value is one , we can also take float values as alpha, rag is from 0 to 1
import matplotlib.pyplot as plt
```

```python
plt.plot([4 , 5 , 6] , [1,3,2] , alpha = 0.5 , linewidth = 10)
```

## zorder  
##### actually it decides that which has to come front and which has to go back , helps in multiple of lines
#### which tells the order of the plot ,,

```python
plt.plot([40,20,30,0,70] ,linewidth = 10 , zorder = 5)
plt.plot([7,4 ,8,0,40] , linewidth = 10 , zorder = 5)
```

```python
plt.plot([7,2,3,0,7] ,linewidth = 10 , zorder = 1)
plt.plot([7,4 ,8,0,4] , linewidth = 10 , zorder = 2)
plt.plot([80,20,30,0,70] ,linewidth = 10 , zorder = 0)
plt.plot([7,40 ,8,0,40] , linewidth = 10 , zorder = 4)
```

#### Setting The Label To The Axis

```python
# x label 
# y label
help(plt.xlabel)
```

```python
# to change the size , font of the label
plt.plot([7,2,3,0,7] ,linewidth = 5 , zorder = 1)
plt.xlabel("x label" , {"size":10 , "color":"b"})
```

```python
# labelpad
plt.plot([1,4,0,7] ,linewidth = 5 , zorder = 1)
plt.xlabel("x label" , {"size":10 , "color":"b"},labelpad = -200,size = 20)
```

```python
#fontdict
# taking the keyword and change the color
plt.plot([7,2,3,0,7] ,linewidth = 5 , zorder = 1)
plt.xlabel("x label" , {"size":10 , "color":"b"}, color = "r")
```

```python
# font properties in x label and y label function
```

```python
plt.plot([1,4,0,7] ,linewidth = 5 , zorder = 1)
plt.xlabel("x label" , size = "xx-small") # x label will be very small
```

```python
# to change the color of the text
plt.plot([1,4,0,7] ,linewidth = 5 , zorder = 1)
plt.xlabel("x label" ,color  = "blue")
plt.ylabel("y label" , color = "yellow")
plt.show()
```

```python
# size , alpha
plt.plot([4,7,8,6,8,10])
plt.xlabel("x label" ,color  = "blue"  ,size = 10 , alpha = 0.4)
plt.ylabel("y label" , color = "yellow", size = 20 , alpha = 1)
# alpha is used for the transparency of the font
```

```python
plt.xlabel("x label" ,color  = "blue"  ,size = 10 , alpha = 0.4 , backgroundcolor = "green")
plt.ylabel("y label" , color = "yellow", size = 20 , alpha = 1 , backgroundcolor = "blue")
```

```python
# to change the font of the font , use style keyword 
plt.xlabel("x label" ,color  = "blue"  ,size = 20 ,style = "italic")
plt.ylabel("y label" , color = "yellow", size = 20 , alpha = 1 , )
```

```python
# use 
help(plt.xlabel)
```

###  Set The Title To The Plot 

```python
# syntax;
# title (label , fontdict = none , loc = none , pad = none , **kwargs)
help(plt.title)
```

```python
plt.title("first  plot" , {"size" : 20 , "color" : "blue"} )
```

```python
# loc
plt.plot([70, 2 , 50 , 55 , 45.5])
plt.title("first  plot" , {"size" : 20 , "color" : "blue"} , loc = "left")
```

```python
# right side # if no value is given it takes as none
plt.plot([70, 2 , 50 , 55 , 45.5])
plt.title("first  plot" , {"size" : 20 , "color" : "blue"} , loc = "right" )
```

```python
# pad  , distance from the text and the axes is differs by using pad funcion
plt.plot([70/2, 2/45 , 50 , 55 , 45.5])
plt.title("first  plot" , {"size" : 20 , "color" : "blue"} ,pad = 20)
```

#### Adding Grid Lines To The Plot

```python
help(plt.grid)
```

```python
plt.grid(b = True)
```

```python
# if nothing in b value it takes as none as default
plt.grid()
plt.grid() # dont use grid to call the function again and again
```

```python
# which parameter
# use minorticks_on() fn to call the minor which function in pylplot
plt.plot([4,5,6,7])
plt.minorticks_on()
plt.grid( b = True , which = "minor")
#Adding Grid Lines To The Plot
```

```python
plt.plot([4,5,6,7])
plt.minorticks_on()
plt.grid(b = True , which = "minor" , color = "blue")
# to change the color of the grid lines
```
